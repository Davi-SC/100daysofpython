```python
import pandas as pd 

cotacao_df = pd.read_csv('./data/indexData.csv')
display(cotacao_df)
```

```python
cotacao_df1 = cotacao_df[['Index','Date','Close']] #diminuindo as colunas
display(cotacao_df1)
```

```python
dfremove= cotacao_df1.loc[(cotacao_df1['Index']!='NYA')] # armazenando os dados que não tem index nya no dfremove
cotacao_dffinal = cotacao_df1.drop(dfremove.index) # removendo os que tinha armazenado no dfremove

cotacao_dffinal = cotacao_dffinal[13900:]# pegando so os ultimos dados
display(cotacao_dffinal.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(cotacao_dffinal['Date'],cotacao_dffinal['Close'],label='NYA',color='g',ls='--',lw='2.5')
plt.legend(loc=2,fontsize=13)
plt.ylabel('Valor fechamento')
plt.xlabel('Data cotação')
plt.title('Cotação da bolsa NYA')
plt.axis('auto')

plt.show()
```
