# Graficos com matplot

```python
#importação

import matplotlib.pyplot as plt
```

```python
# plot() função para gerar grafico e show() para mostrar
x = [1,2,3,4,3,2,1]
y = [2,3,4,2,3,4,2]

plt.plot(x,y)
plt.show()
```

```python
#Personalizando o grafico
plt.plot(x,y,label='Dados')
plt.ylabel('Eixo y')
plt.xlabel('Eixo x')
plt.title("Titulo")
plt.xticks([1,2,3,4])
plt.yticks([1,2,3,4,5])
plt.legend()
plt.show()
```

```python
#Lines propriedades

plt.plot(x,y,linestyle='dotted', color='m', lw =7.0) #na documentação tem mais exemplos
plt.show()
```

```python
plt.scatter(x,y)#scatter gera um grafico de disperção
plt.show()
```

```python
plt.bar(x,y,color='r')#grafico de barras
plt.show()
```

```python
#matplotlib.markers na documentação
plt.scatter(x,y,marker='s', color='y') # s de square
plt.show()
```

```python
# alterando limites do graficco
plt.plot(x,y)
# plt.axis(xmin=-1,xmax=5)
# plt.axis(ymin=1.5,ymax=5)
plt.axis('square')

plt.show()
```

```python
#Figure e Subplots

figura = plt.figure(figsize=(10,3))#figsize trabalha com polegadas
figura.suptitle('Titulo geral')

figura.add_subplot(131)# figura tem uma linha, 3 colunas e o grafico ta na posição 1
plt.plot(x,y,label='dado')
plt.legend()
plt.title('Grafico 1')

figura.add_subplot(132)#agora na posição 2
plt.scatter(x,y,label='dado')
plt.legend()
plt.title('Grafico 2')

figura.add_subplot(133)
plt.bar(x,y,label='dado')
plt.legend()
plt.title('Grafico 3')

plt.savefig('garficosexemplo')# salva a figura no diretorio

plt.show(figura)
```
