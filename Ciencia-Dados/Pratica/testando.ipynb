```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv("car_price_dataset.csv")
display(cars.head(5))
```

```python
#ordenando o dataset pelo menor mileage e owner_count
cars = cars.sort_values(["Mileage","Owner_Count"], ascending=[True,True])
```

```python
display(cars)
```

```python
cars.info()
```

```python
#separando carros eletricos/hibridos dos demais
cars_eletric = cars[cars['Fuel_Type'].str.contains('Electric|Hybrid',case=False)]
cars_fossil = cars[cars['Fuel_Type'].str.contains('Petrol|Diesel',case=False)]
cars_eletric = cars_eletric.sort_values(["Mileage","Owner_Count"], ascending=[True,True])
cars_fossil = cars_fossil.sort_values(["Mileage","Owner_Count"], ascending=[True,True])
print("eletricos e hibridos")
display(cars_eletric.head())
print("carros a combustivel fossil")
display(cars_fossil.head())
```

```python
#tirando algumas informações(doors so para limpar a tabela owner_count)
cars_eletric.drop(columns=['Doors','Owner_Count'],inplace=True)
cars_fossil.drop(columns=['Doors','Owner_Count'],inplace=True)
cars.drop(columns=['Doors','Owner_Count'],inplace=True)
```

```python
#CONSULTANDO
print("Pesquisar carro por:")
print("1 - Marca")
print("2 - Modelo")
print("3 - Ano")
option = input(">> ")

match option:
    case "1":
        pesquisa1 = input("Digite a marca ->  ")
        resultado = cars.query(f'Brand == "{pesquisa1}"')
        display(resultado.sort_values(["Mileage"], ascending=True))
    case "2":
        pesquisa2 = input("Digite o modelo ->  ")
        resultado = cars.query(f'Model == "{pesquisa2}"')
        display(resultado.sort_values(["Mileage"], ascending=True))
    case "3":
        pesquisa3 = input("Digite o ano ->  ")
        resultado =cars.query(f'Year== {pesquisa3}') # em pesquisa por numeros não pode utilizar as aspas na pesquisa
        display(resultado.sort_values(["Mileage"], ascending=True))
    case _:
        print("Opção invalida")
    
```
