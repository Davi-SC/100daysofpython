# Bubble Sort:
 - O algoritmo de ordenação bubble sort funciona comparando dois elementos adjacentes e trocando-os se estiverem na ordem errada.
 - O algoritmo repete esse processo até que todos os elementos estejam na ordem correta.
 - É um algoritmo de ordenação simples, mas não é eficiente para grandes conjuntos de dados.


```python
# exemplo 1 - bubble sort

import random

def bubble(vetor):
    fim = len(vetor)
    while fim > 0:
        j = 0
        while j < fim-1:
            if vetor[j]>vetor[j+1]:
                temp = vetor[j]
                vetor[j] = vetor[j+1]
                vetor[j+1] = temp
            j += 1
        fim -= 1

vetor = list(range(0,10000))
random.shuffle(vetor)
print(vetor)

bubble(vetor)

print(vetor)
```

# Selection Sort
 - O Selection Sort é um algoritmo de ordenação que funciona selecionando o menor elemento da lista e movendo-o para a primeira posição, em seguida, selecionando o segundo menor elemento e movendo-o para a segunda posição, e assim por diante, até que a lista esteja completamente ordenada.
 - Assim como o Bubble sort, o selection sort é um algoritmo de ordenação simples que não é muito eficiente para listas grandes.
 

```python
# exemplo 2 - selection sort
def buscarMenor(vetor): 
    menor = vetor[0]
    menor_indice = 0
    for i in range(1, len(vetor)):
        if vetor[i] < menor:
            menor = vetor[i]
            menor_indice = i
    return menor_indice

def selectionSort(vetor):
    newVetor = []
    i = 0
    while i<len(vetor):
        menor = buscarMenor(vetor)
        newVetor.append(vetor.pop(menor))
        i+=1
    return newVetor

vetor = list(range(0,10000))
random.shuffle(vetor)
print(vetor)
vetor = selectionSort(vetor)
print(vetor)
```

# Insertion Sort
 - O algoritmo de ordenação por inserção trabalha inserindo um elemento em uma posição específica do array, de acordo com o valor do elemento.
 - O algoritmo percorre o array, comparando cada elemento com os elementos anteriores e inserindo-o na posição correta.

```python
# exemplo 3 - insertion sort
def insertion(vetor):
    i = 1
    while i<len(vetor):
        temp = vetor[i]
        trocou = False
        j = i - 1
        while j>=0 and vetor[j]>temp:
            vetor[j+1] = vetor[j]
            trocou = True
            j -= 1
        if trocou:
            vetor[j+1] = temp
        i += 1
    return vetor

#vetor = list(range(0,7))
vetor = [11,4,3,4,25,666,0]
random.shuffle(vetor)
print(vetor)
vetor = insertion(vetor)
print(vetor)
```

# Dividir para conquistar
 - Dividir para conquistar é um paradigma de resolução de problemas que consiste em dividir o problema em subproblemas, resolver cada subproblema recursivamente e, em seguida, combinar as soluções dos subproblemas para obter a solução do problema original.

# Quicksort
 - O quicksort é um algoritmo de ordenação eficiente que funciona dividindo uma lista em duas partes menores, ordenando cada parte separadamente e, em seguida, combinando as partes ordenadas para obter a lista ordenada.
 - O quicksort utiliza dividir para conquistar para ordenar a lista. Ele escolhe um elemento da lista como pivô e divide a lista em duas partes: uma parte com elementos menores que o pivô e outra parte com elementos maiores que o pivô. Em seguida, o quicksort é aplicado recursivamente a cada uma das partes até que toda a lista seja ordenada.

```python
# exemplo 4 - Quick sort

def quicksort(array):
    if len(array) < 2:
        return array
    else: 
        pivo = array[0]
        maiores = []
        menores = []
        for i in range(1, len(array)):
            if array[i] <= pivo:
                menores = menores + [array[i]]
            else:
                maiores = maiores + [array[i]]
    return quicksort(menores) + [pivo] + quicksort(maiores)

vetor = list(range(1,100000))
random.shuffle(vetor)
print(vetor)
vetor = quicksort(vetor)
print(vetor)
```
