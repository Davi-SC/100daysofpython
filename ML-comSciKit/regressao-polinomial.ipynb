# Regressão Polinomial com sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Gerar X com 25 pontos espaçados entre -2 e 2
X = np.linspace(-2, 2, 100)
# Função: y = 0.5*X³ - 1.5*X² + 2*X + 1 + ruído
y = 0.5 * X**3 - 1.5 * X**2 + 2 * X + 1 + np.random.normal(0, 0.5, 100)

dados = pd.DataFrame({
    'x': X,
    'y': y
})
```

```python
dados.head()
```

```python
X = dados['x'].values
Y = dados['y'].values
```

```python
plt.scatter(X,Y,label='Y(X)')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend();
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# reajustes

caracteristicas2 = PolynomialFeatures(degree=2)
X = X.reshape(-1,1)
#Linearizando o X para um formato que a biblioteca compreenda
X_polinomio2 = caracteristicas2.fit_transform(X)
```

```python
modelo2 = LinearRegression()
modelo2.fit(X_polinomio2, y)
Y_polinomio2 = modelo2.predict(X_polinomio2)
```

```python
plt.scatter(X,Y,label='Y(X)')
plt.plot(X,Y_polinomio2,color='red',label='Ajuste')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend();
```

Polinomio de grau 2 não ficou com o ajuste ideal

```python
#Grau 3
caracteristicas3 = PolynomialFeatures(degree=3)
X = X.reshape(-1,1)
X_polinomio3 = caracteristicas3.fit_transform(X)
```

```python
modelo3 = LinearRegression()
modelo3.fit(X_polinomio3, y)
Y_polinomio3 = modelo3.predict(X_polinomio3)
```

```python
plt.scatter(X,Y,label='Y(X)')
plt.plot(X,Y_polinomio2,color='red',label='Grau2')
plt.plot(X,Y_polinomio3,color='black',label='Grau3')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend();
```

O grau 3 teve melhor resultao mas ainda tiveram muitos pontos dispersos

Podemos fazer para graus maiores e calcular o erro medio absoluto para comparar os resultados.
